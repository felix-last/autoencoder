```python
results_filtered = []
for res in results:
#     display(res)
    res_filtered_1 = res.loc[ res['kmeans_n_init'].notnull() ] # get only rows where kmeans_n_init is set
    
#     res_filtered_2 = res.loc[ (res.index.get_level_values('Unnamed: 4') == 'count') & (res['Balanced F-Measure'] == 5) ] # filter where count = 5 to make SDs comparable
#     res_filtered_2.reset_index(-1, inplace=True)
#     #res_filtered_2 = [(a,b,c) for (a,b,c,d) in res_filtered_2.index.get_values()] # remove 'count' because we want access to all 
#     display(res_filtered_1)
#     display(res_filtered_2)
#     res_filtered = res_filtered_1.loc[ res_filtered_2.index.get_values() ]
    res_filtered = res_filtered_1
#     display(res_filtered)
    res_sd = res_filtered.loc[res_filtered.index.get_level_values('Unnamed: 4') == 'std']
    n_init = res_filtered.loc[res_filtered.index.get_level_values('Unnamed: 4') == 'mean'][['kmeans_n_init']]
    res_sd = res_sd.reset_index(level=3,drop=True)
    n_init = n_init.reset_index(level=3,drop=True)
    res_sd.drop(['kmeans_n_init'], axis=1, inplace=True) # SD of n_init is useless
#     res_sd = res_sd.merge(n_init, on=['Dataset','Oversampling','Classifier'])
    res_sd = pd.concat([res_sd,n_init], axis=1)
    display(res_sd)
    res_sd = res_sd.reset_index(level=[0,1], drop=True)
    res_sd.set_index('kmeans_n_init', append=True, inplace=True)
    results_filtered.append(res_sd)
sd_by_n_init = pd.concat(results_filtered, axis=0)
sd_by_n_init
```
